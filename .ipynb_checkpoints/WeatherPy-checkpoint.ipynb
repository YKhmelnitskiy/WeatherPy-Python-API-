```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import csv
# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
#Set our variables
url = "http://api.openweathermap.org/data/2.5/weather?"
Clean_cities = []
Max_temperature = []
Country = []
Date = []
Humidity = []
Cloudiness = []
Lat = []
Lon = []
Wind_speed = []
#setting our units to be imperical (e.g. Fahrenheit, mph)
units = "imperial"
i = 1
y = 1
temperature = []

#Run our `for loop` for city in cities
for City in cities:
#Running a try for each of our functions to find a KeyValue error,if so we don't want that city, and skip it
    try:
        query_url = f"{url}appid={api_key}&q={City}&units={units}"
    
# Get weather data
        weather_response = requests.get(query_url)
        weather_json = weather_response.json()
#Adding our needed variables to separate lists based on the locations in the JSON
        Max_temperature.append(weather_json["main"]["temp_max"]) 
        Country.append(weather_json["sys"]["country"])        
        Cloudiness.append(weather_json["clouds"]["all"])        
        Date.append(weather_json["dt"])        
        Humidity.append(weather_json["main"]["humidity"])        
        Lat.append(weather_json["coord"]["lat"])        
        Lon.append(weather_json["coord"]["lon"])        
        Wind_speed.append(weather_json["wind"]["speed"])
#Adding only the cities that have information      
        Clean_cities.append(City)
#This is where we skip the city
    except KeyError:
        print("City not found. Skipping...")
#Printing our progress with getting our clean data 
    print(f"Processing Record {i} of Set {y}| {City}")
#To show the progress through sets
    if i > 49:
        y = y+1
        i =1
    else:
        i = i+1
        
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
#Making our dataframe through our variables found through the API
City_DataFrame = pd.DataFrame({"City":Clean_cities,
                             "Cloudiness":Cloudiness,
                             "Country":Country,
                             "Date": Date,
                             "Humidity":Humidity,
                             "Lat":Lat,
                             "Lon":Lon,
                             "Max Temp":Max_temperature, 
                             "Wind Speed":Wind_speed
                            }
                             )
```

```python
#Exporting the city data to a .csv
City_DataFrame.to_csv(output_data_file, encoding='utf-8', index=False)
```

#### Latitude vs. Temperature Plot

```python
#Plotting our data with latitude as our x value and 
plt.scatter(Lat,Max_temperature, marker = "o", facecolors="blue",edgecolors="black")

#Setting our limits for the plot (x and y) 
plt.ylim(min(Max_temperature)-10,max(Max_temperature)+10) 
plt.xlim(min(Lat)-10,max(Lat)+10)

#Setting our title and labels
plt.title("City Latitude vs. Max Temperature (03/22/19)")
plt.xlabel("Latitude")
plt.ylabel("Max Temperature (F)")

#Setting a grid for more readibility
plt.grid()

#Saving plot as a figure in my output data folder
plt.savefig("output_data/Lat_v_Max_temp")
```
